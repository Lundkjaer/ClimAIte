```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
electricity = [84480,97622,	114172,	122794,	144384,	169204,	175423,	190727]
gas = [55144,	60662,	71740,	78724,	96479,	115712,	117094,	120291]
dates = ['2021-09-01',	'2021-10-14',	'2021-11-30', '2021-12-23',	'2022-02-22',	'2022-06-10',	'2022-07-12',	'2022-10-05']
dates = pd.to_datetime(dates)
print(dates)
print(type(dates))

df = pd.DataFrame(list(zip(dates, electricity, gas)), columns=['Date', 'Electricity Total kWh', 'Gas Total kWh'])
```

```python
df['Electricity Use kWh'] = df['Electricity Total kWh'].diff(periods=1)
df['Gas Use kWh'] = df['Gas Total kWh'].diff(periods=1)
df.head()
```

```python
plt.figure()
df.plot.line(x='Date', y=['Electricity Use kWh', 'Gas Use kWh'])
plt.ylabel('kWh')
plt.plot(pd.to_datetime(['2022-02-22','2022-06-10']), [10000,10000], marker='|', color='grey')
df.plot.line(x='Date', y=['Electricity Total kWh', 'Gas Total kWh'])
```
