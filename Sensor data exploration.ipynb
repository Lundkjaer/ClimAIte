```python
light_file_list = [x for x in os.listdir() if 'Light.csv' in x]
# os.listdir()
print(light_file_list)
```

```python
custom_date_parser = lambda x: datetime.strptime(x, "%m/%d/%Y %H:%M")

dfs = []
for i in light_file_list:

    df = pd.read_csv(i,
                    parse_dates=['DateTime'],
                    date_parser=custom_date_parser,
                    index_col=['DateTime'])
    dfs.append(df)
```

```python
dfs[1].head()
```

```python
f, axes = plt.subplots(2, 3)
for c, i in enumerate(axes):
  print(c, i)
  for each in i:
    axes[c][each].plot(dfs[c]['Light'])
    axes[c][each].set_title('cats')
# plt.tight_layout()
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15, 6), layout='constrained')
count = 0
for ax, dataframe in zip(axs.flat, dfs):
    y = dataframe['Light']
    ax.set_title(light_file_list[count])
    ax.plot(y)
    count += 1
```
